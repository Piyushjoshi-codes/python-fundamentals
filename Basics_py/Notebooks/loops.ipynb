#Python-Loops Fundamentals
Loops are the iterables in python that allows us to run a specific block of codes multiple times.
Loops are of two types >>>
1) for loops >> For loops are the types of loops in which the condition is to be checked and on that basis, the loop is to be executed. It has an initializer that sets up the loop control variable, condition statement that checks conditions on that basis the loop runs, and update expression that modifies the intialized variable.

2) while loops >> While loop is a type of loops that repeatedly executes a block of code as long as a specified condition remains true. It just has a condition statement that is being checked and if True, then the block of code runs repeatedly and if False then the loop ends.
There's another loop called do-while loop which is a variant of while loop in which the execution of block of codes takes place atleast once before the condition is being checked for the first time.

Examples >>>
1. Using for loop in program >>

```python
list_fruits = ["Apple", "Banana", "Cherry", "Grapes"]
for fruit in list_fruits:     #for item_name in items:
    print(fruit)
```

2. Using for loop with range in program >>

Before forwarding towards the program;
>>> range() function is a built-in function in python in which start_num specifies the starting number to be printed, end_num specifies upto the numbers to be printed but the end_num will not be be printed an at last step number is a number that skips specified amount of of number from start to end..

```python
for i in range(10, -1, -1):        #negative step indicates reverse format.
    print(i)
```

```python
for i in range(1, 10, 3):   #Here it will start printing from 1 and skips 2, 3 and prints 4 and so on...
    print(i)
```

```python
number1 = int(input("Enter a number to print the table of :"))
for i in range(number1, 10 * number1 + number1, number1):     #range(start_num, end_num, step)
    print(i)
```

3. Using while loop in program >>

```python
key = 5
while key != -1:
    print("Countdown : ", key)
    key -= 1
print("Happy New Year !!!")
```

4. using break functions in loops >>

```python
search = [12, 54, 43, 64, 28, 97]
print(search)
key = int(input("Enter the number to be searched from above list : "))
for found in search:
    print("Searching :", found)
    if key == found:
        print (f"{found} was found in the list.")
        break         #after the condition is satisfied it will run once and break function will stop the for loop.
    else:
        print("Number was not found !")
print("Loop Ended!")
```

5. Using continue functions in the loops >>

```python
even_num = int(input("Enter the numbers till you want to print even numbers :"))
for i in range(1, even_num):
    if i % 2 != 0:
        continue       #All the odd numbers are skipped resulting in printing the even numbers.
    print(i)
```
