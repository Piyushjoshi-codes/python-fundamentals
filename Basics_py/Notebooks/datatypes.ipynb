#Python-Datatypes Fundamentals
Data-type -> It is a type data to represent several forms of informaton when processed.
It is a classification in programming that defines the kind of value a variable holds. 
There are several types of data-types in python such as..

***Mutability and Immutability of a data-type:
A data-type variable in which changes can be done after been declared, is said to be Mutable, while a variable that cannot be changed after been declared, is said to be Immutable. We require another variable for performing such changes...

1. String: A string is a collection of several characters, numbers, special characters, etc in a single
or double quotation marks. A string is always immutable.

```python
str_sample = "5 is my favourite number."
print(str_sample)
type(str_sample)
```

2. Float: A float contains decimal point values. It is Immutable.

```python
float_sample = 925.239
print(float_sample)
type(float_sample)
```

3. Integer: An int data-type contains integer values. It is mutable.

```python
int_sample = 780
print(int_sample)
type(int_sample)
```

4. Boolean: A bool data-type contains boolean values which are 'True' or 'False'. 

```python
bool_sample1 = True
bool_sample2 = False
print(bool_sample1 + " " + bool_sample2)
type(bool_sample1)
type(bool_sample2)
```

5. List: A list can hold same or mixed data-types. It is mutable.

```python
list_sample = [1, 2, "apple", "banana"]
print(list_sample)
type(list_sample)
```

6. Tuple: A tuple is an ordered, immutable collection of items. It is immutable.

```python
tuple_sample = 1, 2, 3, 4, "apple"   #A tuple requires round bracket or can be written without it.
print(tuple_sample)
type(tuple_sample)
```

7. Dictionary: A dictionary is a built-in data-type that stores the data as the collection of key-value pairs. It is mostly used for mapping purpose and it is mutable.

```python
dict_sample = {1:"apple", 2:"banana", 3:"Cherry", 4:"Dragon-fruit"}
print(dict_sample)
type(dict_sample)
```
