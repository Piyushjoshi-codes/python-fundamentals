#Python-Datatypes Fundamentals
Data-type -> It is a type data to represent several forms of informaton when processed.
It is a classification in programming that defines the kind of value a variable holds. 
There are several types of data-types in python such as..

***Mutability and Immutability of a data-type:
A data-type variable in which changes can be done after it is been declared, is said to be Mutable, while a variable that cannot be changed after been declared, is said to be Immutable. We require another variable for performing such changes or converting that data-type into a mutable one and then apply changes or many more...

##Numeric data-types >>>

Float: A float contains decimal point values. It is Immutable.

```python
float_sample = 925.239
print(float_sample)
type(float_sample)
```

Integer: An int data-type contains integer values. It is mutable.

```python
int_sample = 780
print(int_sample)
type(int_sample)
```

Complex: A complex data-type that stores both real and imaginary number. This are immutable.

```python
complex_sample = 12 + 5j 
print(complex_sample)
type(complex_sample)
```

##Boolean data-types >>>

Boolean: A bool data-type contains boolean values which are 'True' or 'False'. 

```python
bool_sample1 = True
bool_sample2 = False
print(bool_sample1 + " " + bool_sample2)
type(bool_sample1)
type(bool_sample2)
```

##Sequence data-types >>>

String: A string is a collection of several characters, numbers, special characters, etc in a single
or double quotation marks. A string is always immutable.

```python
str_sample = "5 is my favourite number."
print(str_sample)
type(str_sample)
```

List: A list can hold same or mixed data-types. It is mutable.

```python
list_sample = [1, 2, "apple", "banana"]
print(list_sample)
type(list_sample)
```

Tuple: A tuple is an ordered, immutable collection of items. It is immutable.

```python
tuple_sample = 1, 2, 3, 4, "apple"   #A tuple requires round bracket or can be written without it.
print(tuple_sample)
type(tuple_sample)
```

Range: A range() is an immutable sequence of numbers, often used in loops to create sequence of integers.

```python
range_sample = range(10, 100 + 1, 10)    #range(start, end, step)
print(list(range_sample))         #Converting range to list to view it's data.
type(range_sample)
```

##Set data-types >>>

Set: A mutable unordered collection of unique data elements.

```python
set_sample = {2, 2, 6, 6, 7, "apple", "banana"}   #notice how the data elements are repeated.
print(set_sample)   #While printing notice that each data elements appear once and the order or the index of strings is not guaranteed
type(set_sample)
```

Frozen Set: It's an immutable and hashable version of standard set.

```python
list_sample = [1, 1, 2, "apple", "apple", "banana"]    #same as set.
frozenset_sample = frozenset(list_sample)
print(frozenset_sample)
type(frozenset_sample)
```

##Mapping data-types >>>

Dictionary: A dictionary is a built-in data-type that stores the data as the collection of key-value pairs. It is mostly used for mapping purpose and it is mutable.

```python
dict_sample = {1:"apple", 2:"banana", 3:"Cherry", 4:"Dragon-fruit"}
print(dict_sample)
type(dict_sample)
```

##Binary data-types >>>

Bytes: This is an immutable sequence of bytes.

```python
byte_sample = b"Hello"
print(byte_sample)       #prints the string as it is.
print(byte_sample[0])    #does not prints the 'h' but its ASCII value i.e 72.
print(byte_sample[1])    #does not prints the 'e' but its ASCII value i.e 101.
print(byte_sample[2])    #does not prints the 'l' but its ASCII value i.e 108.
print(byte_sample[3])    #does not prints the 'l' but its ASCII value i.e 108.
print(byte_sample[4])    #does not prints the 'o' but its ASCII value i.e 111.
type(byte_sample)
```

Byte-array: This is a mutable counterpart of bytes.

```python
byteArray_sample = bytearray(b"Hello")
print(byteArray_sample)       #prints the string as it is.
print(byteArray_sample[0])    #does not prints the 'h' but its ASCII value i.e 72.
print(byteArray_sample[1])    #does not prints the 'e' but its ASCII value i.e 101.
print(byteArray_sample[2])    #does not prints the 'l' but its ASCII value i.e 108.
print(byteArray_sample[3])    #does not prints the 'l' but its ASCII value i.e 108.
print(byteArray_sample[4])    #does not prints the 'o' but its ASCII value i.e 111.
type(byteArray_sample)
```

```python
#Another example of byteArray...
bA_sample = bytearray([72, 101, 108, 108, 111, 33]) #prints the string formed from the ASCII codes.
print(bA_sample)
type(bA_sample)
```

Memory-view: It is a mutable object that allows efficient, zero-copy access to the internal data of the other objects such as bytes and byteArrays. 
Zero-copy access- It means it provides the view of original data rather than making it's copy.

```python
byte_sample = b"Hello"
bA_sample = bytearray([72, 101, 108, 108, 111, 33])

#Creating copies of objects supporting buffer protocol.
memoryview_sample1 = memoryview(byte_sample)
memoryview_sample2 = memoryview(bA_sample)
print(memoryview_sample1)  #prints memory address like 0x.....
print(memoryview_sample2)
type(memoryview_sample1)  #gives a single type() 
type(memoryview_sample2)
```

##None data-type >>> 

NoneType: It is a data-type that represents absence of a value or a null value.

```python
Nonetype_sample = None   #Assigned a null value or simply assigned nothing.
print(Nonetype_sample)
type(Nonetype_sample)
```
