While coding sometimes we require specific choices to run some blocks of codes in the program and so for that we use conditions.
Thus, Conditions in python comes with various types such as >>>

1. if condition >> If the conditional statement is true, it will run and execute the code inside the if condition and if false, it will not enter the block 
After the execution is done, it will omit outside the if condition block.

```python
int_sample = 5
if int_sample > 0:      #Condition becomes True and will execute...
    print("Statement is true!") 
if int_sample < 0:      #Condition becomes False and will not execute...
    print("Statement is false")
```

2. if-else condition >> If the conditional statement is true, it will run the block of codes inside the if condition, else it will run the block of codes inside the else condition. It is used only when the condition is based on only two outcomes.

```python
lock = 432874
key = int(input("Enter the password:"))
if key == lock:
    print("Unlocked !")
else:
    print("The Password is Incorrect!")
```

3. if-elif-else condition >> Suppose you have more than one conditions to be checked, it will not be possible with if-else condition. The elif statement can be used to check multiple conditions.

```python
percent = float(input("Enter the percentage marks :"))
if 90 <= percent <= 100:
    print("Brilliant!!")
elif percent >= 80:
    print("Great!")
elif percent >= 65:
    print("Better!")
elif percent >= 35:
    print("Good...")
else:
    print("You failed!")
```
