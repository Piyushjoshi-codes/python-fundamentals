Functions in a programming language are the reusable blocks of codes that performs some specific tasks.
They are useful in python as you don't have to write a ton of code everytime you need it. Functions can be used by first defining it, declaring it and then calling it wherever necessary. 

When we talk about parameters used in function, they are nothing but temporary variables for performing operations on value based programs. They are only used when the function is to be defined and declared.
And about the arguments, they are variables used to actually perform those operations instead of parameters. They are used only when the function is to be called when needed.

Function examples >>>

Function without parameters >>

```python
def greet():
    return "Good Morning"
greet()
```

Function with a single parameter >>

```python
def area(r):
    return f"{3.14 * r * r} m^2"    
radius = float(input("Enter the radius of the circle: "))
print("Area of the cirlce: ")
area(radius)
```

Function with multiple parameters >>

```python
def intro(n, a, c):
    return f"My name is {n} and I'm {a} years old. I want to become {c}."
my_name = input("Enter your name: ")
age = input("Enter your age: ")
career_goal = input("Enter your future profession: ")
intro(my_name, age, career_goal)
```

Function with conditions >>

```python
def is_even(n):
    return n % 2 == 0
number = int(input("Enter the number to check if its an even number: "))
if is_even(number):
    print(f"{number} is an even number.")
else:
    print(f"{number} is an odd number.")
```

Function with a loop >>

```python
def sum_of(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
number = int(input("Enter the number upto you want to print sum of: "))
print(f"Total sum = {sum_of(number)}")
```

Mini-Task Function >>

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
number = int(input("Enter the number to find its factorial: "))
print(f"Factorial of {number} is {factorial(number)}")
```
