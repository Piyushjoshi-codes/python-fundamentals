#Python-Loops Fundamentals
Loops are the iterables in python that allows us to run a specific block of codes multiple times.
Loops are of two types :
1) for loops - For loops are the types of iterables in which the condition is to be checked and on that basis, the loop is to be executed. It has an initializer that sets up the loop control variable, condition statement that checks conditions on that basis the loop runs, and update expression that modifies the intialized variable.

2) while loops - While loop is a type of iterable that repeatedly executes a block of code as long as a specified condition remains true. It just has a condition statement that is being checked and if True, then the block of code runs repeatedly and if False then the loop ends.
There's another loop called do-while loop which is a variant of while loop in which the execution of block of codes takes place atleast once before the condition is being checked for the first time.

Examples :
1. Using for loop in program :

```python
list_fruits = ["Apple", "Banana", "Cherry", "Grapes"]
for fruit in list_fruits:
    print(fruit)
```

2. Using for loop with range in program :

```python
number1 = int(input("Enter a number to print the table of :"))
for i in range(number1, 10 * number1 + number1, number1):
    print(i)
```

3. Using while loop in program : 

```python
key = 5
while key != -1:
    print("Countdown : ", key)
    key -= 1
print("Happy New Year !!!")
```

4. using break functions in loops : 

```python
search = [12, 54, 43, 64, 28, 97]
print(search)
key = int(input("Enter the number to be searched from above list : "))
for found in search:
    print("Searching :", found)
    if key == found:
        print (f"{found} was found in the list")
        break
    else:
        print("Number was not found !")
print("Loop Ended!")
```

```python
list_sam = [12, 23, 99, 76, 54, 45]
search = int(input("Enter the item to be searched : "))
for sam in list_sam:
    print("Checking : ", sam)
    if search == sam:
        print("Item found ")
        break                       #breaks the condition after execution
    else: 
        print("Item not found ")
print("Loop Ended")
```
